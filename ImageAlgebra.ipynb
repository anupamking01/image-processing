```python
import cv2 as cv 
```

```python
#ADDITION
img1 = cv.imread("1.png")
img2 = cv.imread("2.png")
img = cv.add(img1, img2)
print "\n\nmatrice 1\n\n"
print img1

print "\n\nmatrice 2\n\n"
print img2

cv.imshow('image',img)
print "\n\nfinal matrice \n\n"
print img
cv.waitKey(0)
```

```python
#SUBTRACTION
img1 = cv.imread("1.png")
img2 = cv.imread("2.png")
img = cv.subtract(img1, img2)
cv.imshow('image',img)
cv.waitKey(0)
```

```python
#AND
img1 = cv.imread("1.png")
img2 = cv.imread("2.png")
img = cv.bitwise_and(img1, img2)
cv.imshow('image',img)
cv.waitKey(0)
```

```python
#XOR
img1 = cv.imread("1.png")
img2 = cv.imread("2.png")
img = cv.bitwise_xor(img1, img2)
cv.imshow('image',img)
cv.waitKey(0)
```

```python
#OR
img1 = cv.imread("1.png")
img2 = cv.imread("2.png")
img = cv.bitwise_or(img1, img2)
cv.imshow('image',img)
cv.waitKey(0)
```
