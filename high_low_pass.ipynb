```python
# t_img = np.random.randint(0,256,(20,30),dtype='uint8')
t_img = cv2.imread('images/space_view.jpg',cv2.IMREAD_GRAYSCALE)
# t_img = np.ones((10,10),dtype='uint8')

kernel = np.ones((7,7))/49
# kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
r_img = apply_kernel(t_img,kernel)

plt.figure("low pass filter")
plt.subplot(121)
plt.imshow(t_img,cmap='gray')
plt.title("original")

plt.subplot(122)
plt.imshow(r_img,cmap='gray')
plt.title("blurred")
plt.show()
```

```python
t_img = cv2.imread('images/moon.jpg',cv2.IMREAD_GRAYSCALE)
# t_img = np.ones((10,10),dtype='uint8')

kernel = np.ones((3,3))*-1
kernel[1,1] = 8
kernel /= 9
# kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
r_img = apply_kernel(t_img,kernel)

plt.figure("high pass filter")
plt.subplot(131)
plt.imshow(t_img,cmap='gray')
plt.title("original")

plt.subplot(132)
plt.imshow(r_img,cmap='gray')
plt.title("edges")

plt.subplot(133)
sharped_img = np.array(t_img.astype(int)+r_img.astype(int),dtype='int')
sharped_img[sharped_img >255] = 255
plt.imshow(sharped_img,cmap='gray')
plt.title("sharped image")
plt.show()
```
