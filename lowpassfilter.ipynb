```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def main():
    image = np.asarray(Image.open('1.png'))
    h,w,s = image.shape
    img = np.zeros((h,w))
    new_image = np.zeros((h,w))
    for i in range(h):
        for j in range(w):
            img[i,j] = int(0.3*image[i,j,0] + 0.59*image[i,j,1] + 0.11*image[i,j,2])   
    im = plt.imshow(img, cmap='gray')
    plt.show()
    for i in range(h):
        for j in range(w):
            if i==0 or i==h-1 or j==0 or j==w-1:
                new_image[i,j]=img[i,j]
            else:
                new_image[i,j]=(img[i-1,j-1]+img[i-1,j]+img[i-1,j+1]+img[i,j-1]+img[i,j]+img[i,j+1]+img[i+1,j-1]+img[i+1,j]+img[i+1,j+1])/9
    im = plt.imshow(new_image, cmap='gray')
    plt.show()

if __name__=='__main__':
    main()
```
